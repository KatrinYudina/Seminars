__Задание 1.__   
Доказать $(\pmb{AB})^{\top} = \pmb{B}^{\top}\pmb{A}^{\top}$.

_Докозательство_

Предположим $$ A = \begin{pmatrix}
x_{\textit{1}} , x_{\textit{2}} , \dots , x_{\textit{n}}
\end{pmatrix}, B = \begin{pmatrix}
y_{\textit{1}} \\
y_{\textit{2}} \\
\dots \\
y_{\textit{n}}
\end{pmatrix}$$
Тогда
$$С = AB = \begin{pmatrix}
x_{\textit{1}} , x_{\textit{2}} , \dots , x_{\textit{n}}
\end{pmatrix}
\begin{pmatrix}
y_{\textit{1}} \\
y_{\textit{2}} \\
\dots \\
y_{\textit{n}}
\end{pmatrix}
=
x_{\textit{1}}y_{\textit{1}} + x_{\textit{2}}y_{\textit{2}} + \dots + x_{\textit{n}}y_{\textit{n}}$$

Транспонируем каждую матрицу
$$ A^{\top} = \begin{pmatrix}
x_{\textit{1}} \\
x_{\textit{2}} \\
\dots \\
x_{\textit{n}}
\end{pmatrix} , B^{\top} = \begin{pmatrix}
y_{\textit{1}} , y_{\textit{2}} , \dots , y_{\textit{n}}
\end{pmatrix}$$

Отсюда получим

$ С' = B^{\top}A^{\top} = 
y_{\textit{1}}x_{\textit{1}} + y_{\textit{2}}x_{\textit{2}} + \dots + y_{\textit{n}}x_{\textit{n}}$

Следовательно $C = C'$, ч.т.д.

__Определение.__ Матрица $\pmb{M}$ называется ортогональной, если выполняется следующее тождество:
$$
\pmb{M M}^\top = \pmb{M}^\top \pmb{M} = \pmb{I}
$$  
где $\pmb{I}$ - единичная матрица. Другими словами, для ортогональной матрицы верно:  
$$
\pmb{M}^{-1} = \pmb{M}^\top
$$

__Задание 2.__   
$\pmb{A}$ и $\pmb{B}$ ортогональны. Покажите равносильность всех следующих утверждений между собой:

1. $\pmb{AB} = \pmb{BA}$
2. $\pmb{A}^\top \pmb{B} = \pmb{BA}^\top$
3. $\pmb{AB}^\top = \pmb{B}^\top \pmb{A}$
4. $\pmb{A}^\top \pmb{B}^\top = \pmb{B}^\top \pmb{A}^\top$

Подсказка: утверждения равносильные, если от одного к другому можно перейти при помощи набора некоторых операций. Например, здесь можно перейти от первого ко второму, от второго к третьему, от третьего к четвертому и от четвертого к первому.

_Решение_:
1. $\pmb{AB} = \pmb{BA}$
    
    Транспонируем матрицы.
    
    $\pmb{AB}^\top = \pmb{BA}^\top$
    
    Транспонированное произведение матриц равно произведению транспонированных матриц взятых в обратном порядке.
    
    $\pmb{B}^\top \pmb{A}^\top = \pmb{A}^\top \pmb{B}^\top$
    
    Домножим слева на $\pmb{B}$.
    
    $\pmb{B} \pmb{B}^\top \pmb{A}^\top = \pmb{B} \pmb{A}^\top \pmb{B}^\top$
    
    Так как матрицы ортогональны, следовательно при перемножении матрицы на её транспонированную матрицу даёт единичную матрицу. Произведение любой матрицы на единичную = самой матрице.
    
    $\pmb{A}^\top = \pmb{B}\pmb{A}^\top \pmb{B}^\top$
    
    Домножим на $\pmb{B}$ справа.
    
    $\pmb{A}^\top \pmb{B} = \pmb{B}\pmb{A}^\top \pmb{B}^\top\pmb{B}$
    
    $\pmb{A}^\top \pmb{B} = \pmb{B}\pmb{A}^\top$

2. $\pmb{A}^\top \pmb{B} = \pmb{BA}^\top$

    Домножаем справа на $\pmb{A}$
    
    $\pmb{A}^\top \pmb{B} \pmb{A} = \pmb{BA}^\top \pmb{A}$
    $\pmb{A}^\top \pmb{B} \pmb{A} = \pmb{B}$
    
    Домножаем на $\pmb{B}^\top$ справа.
    
    $\pmb{A}^\top \pmb{B} \pmb{A} \pmb{B}^\top = \pmb{B} \pmb{B}^\top$
    $\pmb{A}^\top \pmb{B} \pmb{A} \pmb{B}^\top = \pmb{I}$
    
    Умножаем на $\pmb{A}$ слева.
    
    $\pmb{A} \pmb{A}^\top \pmb{B} \pmb{A} \pmb{B}^\top = \pmb{A}$
    $\pmb{B} \pmb{A} \pmb{B}^\top = \pmb{A}$
    
    Умножаем слева на $\pmb{B}^\top$.
    
    $\pmb{B}^\top \pmb{B} \pmb{A} \pmb{B}^\top = \pmb{B}^\top \pmb{A}$
    $\pmb{A} \pmb{B}^\top = \pmb{B}^\top \pmb{A}$
    
3. 
$\pmb{AB}^\top = \pmb{B}^\top \pmb{A}$
   
   Умножаем на $\pmb{A}^\top$ справа.
   
   $\pmb{AB}^\top \pmb{A}^\top = \pmb{B}^\top \pmb{A} \pmb{A}^\top$
   $\pmb{AB}^\top \pmb{A}^\top = \pmb{B}^\top$
   
   Умножаем на $\pmb{A}^\top$ слева.
   
   $\pmb{A}^\top \pmb{AB}^\top \pmb{A}^\top = \pmb{A}^\top \pmb{B}^\top$
   $\pmb{B}^\top \pmb{A}^\top = \pmb{A}^\top \pmb{B}^\top$
   
   Меняем местами.
   
   $\pmb{A}^\top \pmb{B}^\top = \pmb{B}^\top \pmb{A}^\top$

4. $\pmb{A}^\top \pmb{B}^\top = \pmb{B}^\top \pmb{A}^\top$
    
    Транспонированное произведение матриц равно произведению транспонированных матриц взятых в обратном порядке.
    
    $\pmb{AB} = \pmb{BA}$

__Задание 1.__ 

Дана следующая квадратная матрица:
$$\begin{pmatrix}
2 & -1 &  &  & 0 \\
-1 & 2 & -1 &  & \\
& \ddots & \ddots & \ddots & \\
& & -1 & 2 & -1\\
0& & & -1 & 2
\end{pmatrix}$$  
На главной диагонали значения равны `2`, на диагоналях на 1 выше и на 1 ниже значения равны `-1`. __Все__ остальные значения равны `0`.    
Необходимо создать такую матрицу с помощью `numpy` и без использования циклов.

````python
def tridiagonal_matrix(n):
    z = np.eye(n,n,dtype=int)*2
    rng = np.arange(n-1)
    z[rng, rng+1] = -1
    z[rng+1, rng] = -1
    print(z)
    pass
````

````python
tridiagonal_matrix(10)
````

$$
\sin{(
